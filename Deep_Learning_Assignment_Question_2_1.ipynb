```python
# Deep Learning Assignment - Question 2.1
# Depth of ANNs

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.datasets import mnist
import pandas as pd
import time

# Set random seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Check if GPU is available
print("TensorFlow version:", tf.__version__)
print("GPU Available: ", len(tf.config.list_physical_devices('GPU')) > 0)

# Load and preprocess MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape(-1, 28*28) / 255.0
x_test = x_test.reshape(-1, 28*28) / 255.0
y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test = tf.keras.utils.to_categorical(y_test, 10)

print(f"Training data shape: {x_train.shape}")
print(f"Test data shape: {x_test.shape}")

# Function to create a model with specified depth
def create_model_with_depth(depth):
    model = Sequential()
    model.add(Dense(64, activation='relu', input_shape=(784,)))
    for _ in range(depth - 1):
        model.add(Dense(64, activation='relu'))
    model.add(Dense(10, activation='softmax'))
    model.compile(optimizer=Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Define depths to test
depths = [1, 2, 4, 8, 16]

# Train models with different depths
depth_histories = {}
depth_training_times = {}
depth_params = {}

for depth in depths:
    print(f"\nTraining model with depth={depth}...")
    model = create_model_with_depth(depth)
    print(f"Model with depth={depth}:")
    model.summary()
    depth_params[depth] = model.count_params()
    start_time = time.time()
    history = model.fit(x_train, y_train, epochs=15, batch_size=128, validation_data=(x_test, y_test), verbose=1)
    end_time = time.time()
    training_time = end_time - start_time
    depth_histories[depth] = history.history
    depth_training_times[depth] = training_time
    print(f"Final training accuracy: {history.history['accuracy'][-1]:.4f}")
    print(f"Final validation accuracy: {history.history['val_accuracy'][-1]:.4f}")
    print(f"Training time: {training_time:.2f} seconds")

# Visualize training and validation accuracy
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
for depth in depths:
    plt.plot(depth_histories[depth]['accuracy'], label=f'Depth={depth}', linewidth=2)
plt.title('Training Accuracy vs. Network Depth')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
for depth in depths:
    plt.plot(depth_histories[depth]['val_accuracy'], label=f'Depth={depth}', linewidth=2)
plt.title('Validation Accuracy vs. Network Depth')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Visualize training and validation loss
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
for depth in depths:
    plt.plot(depth_histories[depth]['loss'], label=f'Depth={depth}', linewidth=2)
plt.title('Training Loss vs. Network Depth')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
for depth in depths:
    plt.plot(depth_histories[depth]['val_loss'], label=f'Depth={depth}', linewidth=2)
plt.title('Validation Loss vs. Network Depth')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Compare final performance metrics
depth_metrics = {
    'Depth': [],
    'Final Train Accuracy': [],
    'Final Val Accuracy': [],
    'Training Time (s)': [],
    'Model Parameters': []
}

for depth in depths:
    depth_metrics['Depth'].append(depth)
    depth_metrics['Final Train Accuracy'].append(f"{depth_histories[depth]['accuracy'][-1]:.4f}")
    depth_metrics['Final Val Accuracy'].append(f"{depth_histories[depth]['val_accuracy'][-1]:.4f}")
    depth_metrics['Training Time (s)'].append(f"{depth_training_times[depth]:.2f}")
    depth_metrics['Model Parameters'].append(depth_params[depth])

depth_df = pd.DataFrame(depth_metrics)
print(depth_df)

# Visualize relationship between depth and performance
plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
depths_array = np.array(depths)
val_accuracies = [float(acc) for acc in depth_metrics['Final Val Accuracy']]
plt.plot(depths_array, val_accuracies, 'o-', linewidth=2, markersize=8)
plt.title('Final Validation Accuracy vs. Depth')
plt.xlabel('Network Depth')
plt.ylabel('Validation Accuracy')
plt.grid(True)
plt.xticks(depths_array)

plt.subplot(2, 2, 2)
training_times_array = [float(t) for t in depth_metrics['Training Time (s)']]
plt.plot(depths_array, training_times_array, 'o-', linewidth=2, markersize=8, color='orange')
plt.title('Training Time vs. Depth')
plt.xlabel('Network Depth')
plt.ylabel('Training Time (seconds)')
plt.grid(True)
plt.xticks(depths_array)

plt.subplot(2, 2, 3)
params_array = np.array(depth_metrics['Model Parameters'])
plt.plot(depths_array, params_array, 'o-', linewidth=2, markersize=8, color='green')
plt.title('Model Parameters vs. Depth')
plt.xlabel('Network Depth')
plt.ylabel('Number of Parameters')
plt.grid(True)
plt.xticks(depths_array)

plt.subplot(2, 2, 4)
efficiency = [val_acc / param for val_acc, param in zip(val_accuracies, params_array)]
efficiency = [e * 1e6 for e in efficiency]  # Accuracy per million parameters
plt.plot(depths_array, efficiency, 'o-', linewidth=2, markersize=8, color='purple')
plt.title('Efficiency (Accuracy per Million Parameters)')
plt.xlabel('Network Depth')
plt.ylabel('Efficiency')
plt.grid(True)
plt.xticks(depths_array)

plt.tight_layout()
plt.show()
```
